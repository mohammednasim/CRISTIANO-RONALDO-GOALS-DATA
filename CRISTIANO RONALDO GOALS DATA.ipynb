## Handling missing values

```python
##filling power_of_shot missing vlue with hean
cr7['distance_of_shot'].fillna(value=cr7['distance_of_shot'].mean(), inplace=True)
```

```python
cr7.isnull().sum()
```

```python
#filling type_of_combined_shot with MODE
mode_com = cr7.type_of_combined_shot.value_counts().keys()[0]
print('mode is :', mode_com)
```

```python
cr7.isnull().sum()
```

```python
##filling remaining second with MEDIAN
cr7.remaining_sec.fillna(value=cr7.remaining_sec.median(), inplace=True)
```

```python
##short id eno
cr7.shot_id_number = pd.Series(np.arange(1,cr7.shot_id_number.shape[0]+1))
cr7.isnull().sum()
```

```python
##filling location and locationy
cr7['location_x'].fillna(value=0,inplace=True)
cr7['location_y'].fillna(value=0,inplace=True)
cr7.isnull().sum()
```

```python
#using forward filling method
print('Null value in column home/away before forward fill:',cr7['home/away'].isnull().sum() )
col = ['home/away','lat/lng','team_name','match_id','match_event_id','team_id','remaining_min','knockout_match']
cr7.loc[:,col] = cr7.loc[:,col].ffill()
print('Null value in column home/away after forward fill:',cr7['home/away'].isnull().sum() )
```

```python
# Filling missing values in the 'shot_basics' column based on the 'range_of_shot' value
# Filling missing values in the 'shot_basics' column based on the 'range_of_shot' value
cr7.loc[(cr7['range_of_shot'] == '16-24 ft.') & (cr7['shot_basics'].isnull()), 'shot_basics'] = 'Mid Range'

mask = (cr7['range_of_shot'] == 'Less Than 8 ft.') & (cr7['shot_basics'].isnull())
choices_lt_8ft = ['Goal Area', 'Goal Line']
probabilities_lt_8ft = [0.75, 0.25]
cr7.loc[mask, 'shot_basics'] = cr7.loc[mask, 'shot_basics'].apply(lambda x: np.random.choice(choices_lt_8ft, p=probabilities_lt_8ft) if pd.isnull(x) else x)

mask = (cr7['range_of_shot'] == '8-16 ft.') & (cr7['shot_basics'].isnull())
choices_8_16ft = ['Mid Range', 'Goal Line']
probabilities_8_16ft = [0.65, 0.35]
cr7.loc[mask, 'shot_basics'] = cr7.loc[mask, 'shot_basics'].apply(lambda x: np.random.choice(choices_8_16ft, p=probabilities_8_16ft) if pd.isnull(x) else x)

mask = (cr7['range_of_shot'] == '24+ ft.') & (cr7['shot_basics'].isnull())
choices_24ft = ['Penalty spot', 'Right Corner', 'Left Corner']
probabilities_24ft = [0.987, 0.007, 0.006]
cr7.loc[mask, 'shot_basics'] = cr7.loc[mask, 'shot_basics'].apply(lambda x: np.random.choice(choices_24ft, p=probabilities_24ft) if pd.isnull(x) else x)

mask = (cr7['range_of_shot'] == 'Back Court Shot') & (cr7['shot_basics'].isnull())
choices_back_court = ['Mid Ground Line', 'Penalty Spot']
probabilities_back_court = [0.85, 0.15]
cr7.loc[mask, 'shot_basics'] = cr7.loc[mask, 'shot_basics'].apply(lambda x: np.random.choice(choices_back_court, p=probabilities_back_court) if pd.isnull(x) else x)

# Display the summary of missing values after filling
na_summary = cr7.isnull().sum()
print(na_summary)
```

```python
##filling missing values if both have nan
cr7.shot_basics.fillna(method='ffill',inplace=True)
cr7.range_of_shot.fillna(method='ffill',inplace=True)
cr7.isnull().sum()
```

```python
##filling missing value in area of shot
cr7.area_of_shot.fillna(value='Center(C)', inplace=True)

cr7.isnull().sum()
```
